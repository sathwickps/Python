```python
import matplotlib.pyplot as plt
import numpy as np

# Bar graph
x = ['A', 'B', 'C', 'D', 'E']
y = [15, 7, 12, 9, 6]

plt.bar(x, y, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph')
plt.show()
```

```python
# Pie chart

sizes = [30, 25, 20, 25]
colors = ['yellow', 'green', 'orange', 'red']

plt.pie(sizes, labels=colors, colors=colors)
plt.title('Pie Chart')
plt.show()
```

```python
# Scatter plot
x = np.random.randn(100)
y = np.random.randn(100)

plt.scatter(x, y, color='skyblue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
```

```python
# Generate random data
data = np.random.normal(1000)
data1 = np.random.normal(2000)
data2 = np.random.normal(5000)
dataset=[data1,data,data2]
```

```python
# Create a histogram
plt.hist(data)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# Create a box plot
plt.boxplot(dataset)
plt.xlabel('Data')
plt.title('Box Plot')
plt.show()
```
